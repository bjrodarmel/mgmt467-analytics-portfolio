### Q1 ‚Äî Exploratory vs. Explanatory
**Prompt:** *"Explain the difference between exploratory and explanatory visualization."*  
**Hints:** Consider audience, purpose, level of polish, and how much iteration you expect.

üìù **Your response:**

- Exploratory visualization is like a private brainstorming session üß†. The audience is usually the creator themselves or a small team. Its purpose is to find patterns, anomalies, and relationships within a dataset. The level of polish is minimal, and it's highly iterative, involving creating many quick charts to test hypotheses and uncover new insights.
- Explanatory visualization is a well-rehearsed presentation üó£Ô∏è. The audience is an external group, like stakeholders, clients, or the public. Its purpose is to communicate a specific, pre-determined insight or "story" found during the exploratory phase. The level of polish is high, with clear titles, annotations, and a clean design. It's not iterative; the final visual is designed to guide the audience to a specific conclusion.
- In short: Exploratory visuals are for discovery, helping the analyst understand the data, while explanatory visuals are for communication, helping the audience understand the analyst's findings. The transition from one to the other marks the shift from analysis to storytelling. üìö

### Q2 ‚Äî Chart Types & When to Use Them
**Prompt:** *"Describe 5 common chart types (e.g., bar, line, scatter, histogram, map) and the type of data analysis question each is best suited to answer."*  

For each chart, list: **Question archetype**, **Data shape**, **Pitfalls**, **Alternatives**.

üìù **Your response (template):**
**Bar Chart**

Question archetype: Comparison. Best for comparing categorical data. (e.g., "Which region had the highest sales?")

Data shape: One categorical and one quantitative variable.

Pitfalls: Misleading if the y-axis doesn't start at zero. Cluttered with too many categories.

Alternatives: Line chart for time series, pie chart for parts of a whole, or a Lollipop chart for cleaner visualization with many categories.

**Line Chart**

Question archetype: Trend. Ideal for showing change over a continuous period. (e.g., "How have monthly website visits changed?")

Data shape: One continuous (typically time) and one or more quantitative variables.

Pitfalls: Implies continuity, making it a poor choice for sparse data. Too many lines create a "spaghetti chart."

Alternatives: A bar chart for discrete time points, or a radial chart for cyclical data.

**Scatter Plot**

Question archetype: Relationship. Used to explore the correlation between two quantitative variables. (e.g., "Is there a link between marketing spend and sales?")

Data shape: Two quantitative variables.

Pitfalls: Overplotting in large datasets can hide patterns. Only shows correlation, not causation.

Alternatives: A 2D histogram or hexbin plot to handle overplotting. A bubble chart to add a third variable.

**Histogram**

Question archetype: Distribution. Best for showing the frequency distribution of a single quantitative variable. (e.g., "What is the age distribution of our customers?")

Data shape: A single quantitative variable, binned into intervals.

Pitfalls: The visual can be heavily altered by the chosen bin size. Often confused with a bar chart, but shows continuous data.

Alternatives: A box plot or violin plot for showing outliers and data spread.

**Map**

Question archetype: Spatial patterns. Excellent for visualizing data across a geographic area. (e.g., "Where are our customers geographically concentrated?")

Data shape: Geographic data (e.g., coordinates, regions) and a quantitative variable.

Pitfalls: Larger areas can appear more important. Density issues in urban areas make it hard to see patterns.

Alternatives: A bar chart to compare regions without geographic bias. A heat map or density map for dense, point-based data.

### Key Performance Indicators for an Airline Dashboard

Here are 5 key performance indicators (KPIs) for an airline executive's dashboard:

***

### KPI 1: On-Time Performance (OTP)
* **Why it matters:** A key indicator of operational efficiency and customer satisfaction. Higher OTP often leads to better customer retention and a stronger brand reputation.
* **Best chart:** A **gauge chart** üìà to show the current on-time percentage against a target. A **line chart** can show the trend of OTP over time.
* **Drill-down fields:** By airline, by route, by time of day, by airport, or by weather conditions.

***

### KPI 2: Load Factor
* **Why it matters:** This measures how full an airline's planes are, representing the percentage of available seats that are filled. It's a critical measure of profitability, as each empty seat is a lost revenue opportunity.
* **Best chart:** A **bar chart** üìä to compare the load factor across different routes or aircraft types. A **line chart** can track the load factor trend over a month or year.
* **Drill-down fields:** By route, by aircraft type, by day of the week, or by time of year (e.g., peak vs. off-season).

***

### KPI 3: Revenue per Available Seat Mile (RASM)
* **Why it matters:** This is a core metric of financial performance, measuring the total revenue generated for every mile flown by every available seat. Executives use it to gauge pricing strategy and market demand.
* **Best chart:** A **line chart** üìà to show the trend of RASM over time and compare it against the Cost per Available Seat Mile (CASM). A **bar chart** can compare RASM across different routes.
* **Drill-down fields:** By route, by fare class (e.g., economy, business), by time of year, or by marketing campaign.

***

### KPI 4: Cancellation Rate
* **Why it matters:** High cancellation rates can lead to significant costs, including re-booking fees, lodging for stranded passengers, and damage to brand reputation. It signals operational issues that need to be addressed.
* **Best chart:** A **bar chart** üìâ to compare cancellation rates across airports or reasons for cancellation. A **line chart** can show the cancellation rate trend over time to identify problematic periods.
* **Drill-down fields:** By airport, by weather, by mechanical issue, or by crew availability.

***

### KPI 5: Net Promoter Score (NPS)
* **Why it matters:** This is a measure of customer loyalty and satisfaction, derived from a single question: "On a scale of 0-10, how likely are you to recommend us to a friend?" It's a leading indicator of future revenue and growth.
* **Best chart:** A **pie chart** ü•ß or **stacked bar chart** to show the breakdown of promoters, passives, and detractors. A **line chart** can track the overall NPS score trend over time.
* **Drill-down fields:** By flight route, by customer segment (e.g., business vs. leisure), by time of year, or by customer feedback.

### Create a Synthetic Airline Dataset (if you don't have one)
Use this dataset to prototype visualizations and dashboards. It contains realistic columns for the KPIs you defined. You can always swap in your BigQuery data later.


````python
# Generate a synthetic airline dataset
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

np.random.seed(467)

dates = pd.date_range(start='2024-01-01', end='2024-06-30', freq='D')
carriers = ['AA', 'DL', 'UA', 'WN', 'AS']
origins = ['ATL','DFW','DEN','ORD','LAX','SEA','SFO','PHX','CLT','MCO']
dests =   ['JFK','LGA','BOS','SFO','LAX','SEA','MIA','IAH','IAD','PHL']

rows = []
for d in dates:
    for carrier in carriers:
        # number of flights per day per carrier
        n = np.random.randint(80, 140)
        for _ in range(n):
            o = np.random.choice(origins)
            de = np.random.choice(dests)
            if o == de:
                continue
            distance = np.random.randint(300, 2800)
            seats = np.random.choice([90,120,150,180,210])
            load_factor = np.clip(np.random.normal(0.82, 0.07), 0.55, 0.98)
            pax = int(seats * load_factor)
            dep_delay = int(np.random.normal(8, 15))
            arr_delay = dep_delay + int(np.random.normal(0, 12))
            dep_delay = max(dep_delay, -5)
            arr_delay = max(arr_delay, -10)
            cancelled = np.random.rand() < 0.025
            diverted = np.random.rand() < 0.005
            revenue = pax * np.random.uniform(120, 420)
            casm = np.random.uniform(0.08, 0.16)  # cost per available seat mile
            cost = casm * seats * distance
            fuel_burn = distance * np.random.uniform(3.0, 4.5)  # gallons per seat-mile proxy
            nps = np.random.normal(35, 15)  # if you want a customer metric proxy
            rows.append({
                'date': d,
                'carrier': carrier,
                'origin': o,
                'dest': de,
                'distance_miles': distance,
                'seats': seats,
                'passengers': pax,
                'dep_delay_min': dep_delay,
                'arr_delay_min': arr_delay,
                'cancelled': int(cancelled),
                'diverted': int(diverted),
                'revenue_usd': round(revenue, 2),
                'cost_usd': round(cost, 2),
                'fuel_burn_gal': round(fuel_burn, 1),
                'nps': round(nps, 1)
            })

df_flights = pd.DataFrame(rows)

# Derived KPIs
df_flights['on_time'] = (df_flights['arr_delay_min'] <= 15).astype(int)
df_flights['load_factor'] = (df_flights['passengers'] / df_flights['seats']).round(3)
df_flights['yield_usd_per_pax_mile'] = (df_flights['revenue_usd'] / (df_flights['passengers'] * df_flights['distance_miles']).clip(lower=1)).round(4)
df_flights['asm'] = df_flights['seats'] * df_flights['distance_miles']  # available seat miles
df_flights['casm'] = (df_flights['cost_usd'] / df_flights['asm']).round(4)
df_flights['month'] = df_flights['date'].dt.to_period('M').astype(str)

df_flights.head()
````

Prompt A Results

````python
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Compute On-Time Rate ---
# Group by carrier, calculate the mean of 'on_time', and sort
on_time_rate = df_flights.groupby('carrier')['on_time'].mean().sort_values(ascending=False)

# --- 2. Create the Plot ---
plt.style.use('seaborn-v0_8-whitegrid') # Use a clean style
fig, ax = plt.subplots(figsize=(12, 7))

# Create the bar plot
barplot = sns.barplot(
    x=on_time_rate.index,
    y=on_time_rate.values,
    palette='viridis',
    ax=ax,
    hue=on_time_rate.index, # Assign hue to avoid the warning
    legend=False
)

# --- 3. Add Percentage Labels ---
# Iterate over the bars and add text labels
for p in barplot.patches:
    ax.annotate(f'{p.get_height():.1%}', # Format as percentage
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                fontsize=11,
                xytext=(0, 9), # Offset text above the bar
                textcoords='offset points')

# --- 4. Customize and Tidy Up ---
ax.set_title('On-Time Arrival Rate by Carrier', fontsize=16, pad=20)
ax.set_xlabel('Carrier', fontsize=12)
ax.set_ylabel('On-Time Rate (%)', fontsize=12)

# Format y-axis to show percentages
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.0%}'))

# Adjust y-axis limit to make space for labels
ax.set_ylim(0, on_time_rate.max() * 1.15)

plt.tight_layout() # Ensure all labels fit
plt.show()
````

````python
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Aggregate Data ---
# Group by date and calculate the mean arrival delay
daily_delay = df_flights.groupby('date')['arr_delay_min'].mean()

# Calculate the 7-day rolling average
rolling_avg_delay = daily_delay.rolling(window=7).mean()

# --- 2. Create the Plot ---
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize=(14, 7))

# Plot the daily average delay
ax.plot(daily_delay.index, daily_delay.values, label='Daily Average Delay', color='lightblue', alpha=0.8)

# Plot the 7-day rolling average
ax.plot(rolling_avg_delay.index, rolling_avg_delay.values, label='7-Day Rolling Average', color='darkgreen', linewidth=2)

# --- 3. Customize and Tidy Up ---
ax.set_title('Average Arrival Delay Over Time', fontsize=16, pad=20)
ax.set_xlabel('Date', fontsize=12)
ax.set_ylabel('Average Arrival Delay (Minutes)', fontsize=12)
ax.legend()
ax.grid(True, which='both', linestyle='--', linewidth=0.5)

# Add a horizontal line at y=0 for reference
ax.axhline(0, color='red', linestyle='--', linewidth=1, label='On-Time')

plt.tight_layout()
plt.show()
````

Prompt C Results

````python
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Calculate Annotation Data ---
# Calculate the percentage of flights within the [-5, 15] minute window
on_time_window_pct = ((df_flights['arr_delay_min'] >= -5) & (df_flights['arr_delay_min'] <= 15)).mean()

# --- 2. Create the Plot ---
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize=(12, 7))

# Plot the histogram
# We'll cap the view to a reasonable range like -30 to 90 to focus on the main distribution
sns.histplot(df_flights['arr_delay_min'].clip(-30, 90), bins=50, kde=True, ax=ax, color='skyblue')

# --- 3. Add Annotations and Reference Lines ---
# Add a vertical line at 0 for a perfect on-time arrival
ax.axvline(0, color='red', linestyle='--', linewidth=1.5, label='On-Time (0 min)')

# Add a vertical line at 15 for the official on-time cutoff
ax.axvline(15, color='orange', linestyle=':', linewidth=1.5, label='FAA On-Time Cutoff (15 min)')

# Annotate the percentage of flights in the window
ax.text(0.95, 0.90, f'{on_time_window_pct:.1%} of flights arrived\nwithin -5 to +15 minutes',
        transform=ax.transAxes, # Use axis coordinates for placement
        fontsize=12,
        verticalalignment='top', horizontalalignment='right',
        bbox=dict(boxstyle='round,pad=0.5', fc='aliceblue', alpha=0.8))

# --- 4. Customize and Tidy Up ---
ax.set_title('Distribution of Flight Arrival Delays', fontsize=16, pad=20)
ax.set_xlabel('Arrival Delay (Minutes)', fontsize=12)
ax.set_ylabel('Number of Flights', fontsize=12)
ax.legend()

plt.tight_layout()
plt.show()
````

Prompt D Results

````python
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Compute Data ---
# Create a 'route' column for easier grouping
df_flights['route'] = df_flights['origin'] + '‚Äì' + df_flights['dest']

# Group by route, calculate mean load factor, and get the top 15
top_15_routes = df_flights.groupby('route')['load_factor'].mean().nlargest(15)

# Sort for plotting (so the highest is at the top of the horizontal chart)
top_15_routes = top_15_routes.sort_values(ascending=True)

# --- 2. Create the Plot ---
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize=(12, 8))

# Create the horizontal bar plot
barplot = sns.barplot(
    x=top_15_routes.values,
    y=top_15_routes.index,
    palette='plasma',
    ax=ax,
    hue=top_15_routes.index,
    legend=False
)

# --- 3. Add Percentage Labels ---
# Add data labels to the end of the bars
for p in barplot.patches:
    width = p.get_width()
    ax.text(width + 0.005, # Position text slightly after the bar
            p.get_y() + p.get_height() / 2,
            f'{width:.1%}', # Format as percentage
            ha='left', va='center',
            fontsize=10)

# --- 4. Customize and Tidy Up ---
ax.set_title('Top 15 Routes by Average Load Factor', fontsize=16, pad=20)
ax.set_xlabel('Average Load Factor (%)', fontsize=12)
ax.set_ylabel('Route', fontsize=12)

# Format x-axis to show percentages
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.0%}'))

# Adjust x-axis limit to make space for labels
ax.set_xlim(right=top_15_routes.max() * 1.1)

plt.tight_layout()
plt.show()
````
